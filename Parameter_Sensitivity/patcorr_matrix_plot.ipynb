# Read pattern correlation matrix and plot as an image

Read in patcorr matrix file (.txt)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axis as ax
```

```python
patcorr_mat = np.loadtxt("patcorr/sum.txt")
```

```python
patcorr_mat.shape
```

Plot image

```python
fig, ax = plt.subplots(figsize=(12,12))
heatmap = ax.imshow(patcorr_mat, cmap='cividis')
#plt.xticks(np.arange(0,34,3),np.arange(1,35,3)) # labels 1-34, every third
#plt.yticks(np.arange(0,34,3),np.arange(1,35,3))
plt.xticks(np.arange(0,34,1),np.arange(1,35,1)) # labels 1-34, every one
plt.yticks(np.arange(0,34,1),np.arange(1,35,1))
ax.xaxis.tick_top() # move x ticks to top
plt.ylabel('Parameter Number')
plt.xlabel('Parameter Number')
ax.xaxis.set_label_position('top') # move x label to top
ax.set_ylim(34-0.5, -0.5) # matplotlib bug fix for y labels
# add colorbar axes with [left, bottom, width, height] 
cax = fig.add_axes([0.93, 0.13, 0.04, 0.74])
cbar = ax.figure.colorbar(heatmap, cax=cax)
#fig.savefig("patcorr_matrix_heatmap.pdf")
```
